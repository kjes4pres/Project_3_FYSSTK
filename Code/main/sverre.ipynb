```python
path = '/uio/hume/student-u15/sverremj/Master/Project_3_FYSSTK/Project_3_FYSSTK/kagglehub/datasets/nikhil7280/weather-type-classification/versions/1/weather_classification_data.csv'
dataset = WeatherDataset(csv_file=path)

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
    
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])
```

```python
batch_size = 64

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader   = DataLoader(test_dataset, shuffle=False)

input_dim = dataset.X.shape[1]  # number of features
```

```python
Network_w = WeatherNN(input_dim=input_dim, hidden_dim=64, num_hidden_layers=3, output_dim=4, activation="sigmoid")
```

```python
Network_w.get_model()
```

```python
for i in range(50):
    f_sum = 0

    for batch in train_loader:
        f = Network_w.train(batch[0], batch[1], lr=0.001)
        f_sum += f
    print(f_sum)
```

```python
total = 0
correct = 0
with torch.no_grad():
    for i in test_loader:
        y_pred = Network_w.pred(i[0])
        y_pred_label = torch.argmax(y_pred)
        correct += (y_pred_label == i[1]).sum().item()
        total += i[1].size(0)
```
